```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##  Read the file

```python
 df=pd.read_csv("employee/myexcel - myexcel.csv.csv")
df
```

## Generate random height values

```python
df['height'] = np.random.randint(150, 180, size=len(df))
df
```

## Qn 1)How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_count=df["Team"].value_counts()
team_count
```

```python
percentage=(team_count/len(df))*100
percentage
```

## Qn 2)Segregate the employees w.r.t different positions.

```python
position_count=df["Position"].value_counts()
position_count
```

## Qn 3)Find from which age group most of the employees belong to.

```python
age_count=df["Age"].value_counts()
age_count
```

```python
common_age=age_count.idxmax()
common_age
```

## Qn 4)Find out under which team and position, spending in terms of salary is high.

```python
spending_salary=df.groupby(["Team","Position"])["Salary"].sum()
spending_salary
```

```python
high_spending_salary=spending_salary.idxmax()
high_spending_salary
```

## Qn 5)Find if there is any correlation between age and salary , represent it visually.

```python
correlation_coefficient=df["Age"].corr(df["Salary"])
correlation_coefficient
```

```python
plt.scatter(df["Age"],df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

## Detailed report and explanation of  employees in each team.

```python

grouped_data = df.groupby('Team')

# Iterate through each team
for team, team_data in grouped_data:
    print("Team:", team)
    
    
    employee_details = team_data[[ 'Name','Number', 'Position', 'Age','Height','Weight','Salary']]
    employee_count = len(employee_details)
    employee_percentage = (employee_count / len(df)) * 100
    
    print("Number of employees:", employee_count)
    print("Percentage of employees:",round(employee_percentage,3),'%')
    print("Employee Details:")
    print(employee_details)
    print("\n")
```
